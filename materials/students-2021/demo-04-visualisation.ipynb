### **2.1 The Matplotlib Package**

+ Library of software used to create graphics and other visual representations 
<p>

```
import matplotlib.pyplot as plt
```

### **2.2 -Creating graphs on Matplotlib** 

+ Simple task which can be done by recovering some of the things we have already learned (e.g.create lists) 



```
# Step 1 - Import all the packages necessary to execute the code. In this case the abbreviation of the matplotlib function is plt 
import matplotlib.pyplot as plt

# Step 2 - Define the x-axis and corresponding y-axis values as lists 

x= [1, 2, 3, 4, 5, 6]
y= [1,4,9,16,25,36]

# Step 3 - Apply the plt.plot (variable1,variable2)command. This will automatically create a 2D axis for your variables
plt. plot (x,y)

## Step 4 - Apply the plt.show()command. This will display the plot you have created in a separate dialogue box 
plt.show()
```

### **2.3 - Mastering your graphs with Matplot**

You can create multiple plots for as many variables as you want <p>
 You can also manipulate the appearance of your graphs (color,markers and line styles), format the axes, add grids,titles,labels and legends.

```
#Try to run this code and notice the differences compared to the previous one apart from the values 
import matplotlib.pyplot as plt

values = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]

ax = plt.axes()
ax.grid()

plt.xlabel('Values')
plt.ylabel('Entries')
plt.title('Evolution of Values')

plt.plot(range(1,11), values,'y--o')
plt.legend(['First'],loc=4)

plt.show()
```

##**3.Creating Graphs Using Pandas**

However, in most cases we won't be working with simples sets of data but rather large ones 

Pandas is software library written for the Python programming language for data manipulation and analysis.

For the following demonstration I will be using data taken from the 2018's European Social Survey and my variable will be ["State of health services in the country nowaday" (stfhlth)](http://nesstar.ess.nsd.uib.no/webview/index.jsp?v=2&submode=variable&study=http%3A%2F%2F129.177.90.83%3A-1%2Fobj%2FfStudy%2FESS9e03.1&gs=undefined&variable=http%3A%2F%2F129.177.90.83%3A80%2Fobj%2FfVariable%2FESS9e03.1_V103&mode=documentation&top=yes) 

```
#Step 1 - Import all the packages necessary to execute the code
import pandas as pd

# Step 2 - Load the data you want to analyze

my_link = 'https://drive.google.com/uc?id=1rmP5UXogG0JavNFSsxiEFNxEO4LXbZPY'
df = pd.read_csv (my_link)

#Step 3 - Apply the Data Frame command (df)
df
```

If you observe the Data Frame you have loaded, you notice two things:

+ **Certain columns don't matter (e.g.round of the survey)**
+ **The size of the simple is massive**








```
#Let's try to select the only column that matters (stfhlth) 
my_column= ['stfhlth']
df[my_column]
```

+ To view all the observations in the table you can use **values_count()** method

```
df[my_column].value_counts()
```


+ In case you want to visualize your categories in order just use **sort_index()**
+ Each method that is inserted must be separated by a period **( . )**

```
df[my_column].value_counts().sort_index()
```

+ Finally, to create your graph by just adding **plot.bar()**

```
df[my_column].value_counts().sort_index().plot.bar()
```

There is many other visual representations you can create using Pandas such as scatter plots, boxplots, histograms, pie charts, etc.
